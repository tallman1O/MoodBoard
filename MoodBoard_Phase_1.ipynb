```python
import numpy as np
import pandas as pd
import glob
```

Datasets :
1. photos.tsv
2. collections.tsv
3. colors.tsv
4. conversions.tsv
5. keywords.tsv


```python
#Loading the datasets in Pandas
path = './'
documents = ['photos', 'keywords', 'collections', 'conversions', 'colors']
datasets = {}

for doc in documents:
  files = glob.glob(path + doc + ".tsv*")

  subsets = []
  for filename in files:
    df = pd.read_csv(filename, sep='\t', header=0)
    subsets.append(df)

  datasets[doc] = pd.concat(subsets, axis=0, ignore_index=True)
```

Merge the datasets one by one on the basis of a primary key photo_id

```python
#Merge photos and keywords dataset.
photos_keywords = datasets['photos'].merge(
    datasets['keywords'], on='photo_id', how='left'
)
```

```python
photos_keywords_colors = photos_keywords.merge(
    datasets['colors'], on='photo_id', how='left'
)
```

```python
photos_keywords_colors_collections = photos_keywords_colors.merge(
    datasets['collections'], on='photo_id', how='left'
)
```

```python
final_combined_dataset = photos_keywords_colors_collections.merge(
    datasets['conversions'], on='photo_id', how='left'
)
```
